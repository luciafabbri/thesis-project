```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Caricare il file CSV
file_path = "./responses_25mar_copia.csv"  # Cambia il nome del file se necessario
df = pd.read_csv(file_path)
df
```

```python
df.describe()
```

```python
#analisi partecipanti 
age_group_counts = df['age_group'].value_counts()
gender_counts = df['gender'].value_counts()

age_group_counts, gender_counts
```

```python
# Conta le occorrenze per 'age_group' e 'gender'
age_group_counts = df['age_group'].value_counts()
gender_counts = df['gender'].value_counts()

# Crea un grafico a barre per le due variabili
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Grafico per la distribuzione dell'età
ax[0].bar(age_group_counts.index, age_group_counts.values, color='lightblue')
ax[0].set_title('Distribuzione per Fasce di Età')
ax[0].set_xlabel('Fasce di Età')
ax[0].set_ylabel('Numero di Partecipanti')

# Grafico per la distribuzione per genere
ax[1].bar(gender_counts.index, gender_counts.values, color='lightgreen')
ax[1].set_title('Distribuzione per Genere')
ax[1].set_xlabel('Genere')
ax[1].set_ylabel('Numero di Partecipanti')

# Mostra i grafici
plt.tight_layout()
plt.show()
```

```python
domande = []
```

```python
# Seleziona solo le colonne che contengono 'qX_text'
domande_colonne = [col for col in df.columns if 'q' in col and 'text' in col]
domande_colonne
```

```python
# Conta il numero totale di commenti (valori non nulli) in tutte le colonne selezionate
totale_commenti = df[domande_colonne].notna().sum().sum()
totale_commenti
```

```python
# Aggiungi i testi delle domande
for col in domande_colonne:
    domande.extend(df[col].dropna())  # Aggiunge le domande alla lista, ignorando i valori NaN
```

```python
# Conta le occorrenze di ciascun testo nelle domande
domande_counts = pd.Series(domande).value_counts()

domande_df = pd.DataFrame({
    'question': domande_counts.index,
    'counts': domande_counts.values
})
```
