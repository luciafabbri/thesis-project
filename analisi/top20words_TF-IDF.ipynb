```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

### TOP 10 parole per ogni tipologia di tossicità

#### 0 - tossicità generica

```python
df = pd.read_csv(
    "./datasets/dataset_tipo_0_toxicity.csv",
    sep=",",
    quoting=csv.QUOTE_MINIMAL,
    encoding="utf-8",
    engine="python"
)
```

```python
# Rimuovo le righe con NaN in comment_cleaned   
df = df.dropna(subset=["comment_cleaned"])  
```

```python
vectorizer = TfidfVectorizer(max_features=20)  
X_tfidf = vectorizer.fit_transform(df["comment_tokenized"])

feature_names = vectorizer.get_feature_names_out()

with open("./topwords/top20words_toxicity.txt", "w", encoding="utf-8") as f:
    for word in feature_names:
        f.write(word + "\n")
```

```python
word_scores = np.asarray(X_tfidf.sum(axis=0)).ravel()
sorted_indices = word_scores.argsort()[::-1]

N = 20
top_words = [(feature_names[i], word_scores[i]) for i in sorted_indices[:N]]
print("Top parole globali:", top_words)

words, scores = zip(*top_words)
plt.figure(figsize=(10, 6))
plt.barh(words, scores)
plt.xlabel("TF-IDF Score")
plt.title(f"Top {N} parole nel corpus")
plt.gca().invert_yaxis()
plt.show()
```
